```python
soup=BeautifulSoup(webpage,'html.parser')
```

```python
soup
```

```python
company=soup.find_all('div',class_="_2kHMtA")
```

```python
company
```

```python

name=[] 
for i in company:
        try:
            name.append(i.find('div',class_="_4rR01T").text.strip())
        except:
            name.append(np.nan)
            

Display=[]
for ultag in soup.find_all('ul', {'class': '_1xgFaf'}):
    for litag in ultag.find_all('li',class_="rgWa7D")[1]:
        Display.append(litag.text)
      
        
ram=[]
for ultag in soup.find_all('ul', {'class': '_1xgFaf'}):
    for litag in ultag.find_all('li',class_="rgWa7D")[0]:
        ram.append(litag.text)
        
Camera=[]
for ultag in soup.find_all('ul', {'class': '_1xgFaf'}):
    for litag in ultag.find_all('li',class_="rgWa7D")[2]:
        Camera.append(litag.text)
        
        
Battery=[]
for ultag in soup.find_all('ul', {'class': '_1xgFaf'}):
    for litag in ultag.find_all('li',class_="rgWa7D")[3]:
        Battery.append(litag.text)
        
Processor=[]
for ultag in soup.find_all('ul', {'class': '_1xgFaf'}):
    for litag in ultag.find_all('li',class_="rgWa7D")[4]:
        Processor.append(litag.text)
new=[]
for i in Processor:
    temp=[]
    count=0
    temp=i.split()
    for j in temp:
        if j=='Warranty':
            count=count+1 
    if count==0:
        new.append(i)
    else:
        new.append('nan')

price=[] 
for i in company:
        try:
            price.append(i.find('div',class_="_1_WHN1").text.strip())
        except:
            price.append(np.nan)
        

        
```

```python
df=pd.DataFrame({'name':name,
   'Display':Display,
   'ram':ram,
   'Battery':Battery,
    'Processor':new,
    'Camera':Camera,
    'price':price
   })
  
```
